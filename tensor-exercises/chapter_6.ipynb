# Reduction Operations

```python
TENSOR_1 = torch.randint(0,1000,(3,2,3))
TENSOR_1
```

```python
# tensor.sum()

print(TENSOR_1)
print(TENSOR_1.sum()) # sum all values of a tensor
print(TENSOR_1.sum(dim=0)) # Sum across rows (vertical). One sum per column.
print(TENSOR_1.sum(dim=1, keepdim=True))   # Sum across columns (horizontal). One sum per row.

# keepdim=True tells PyTorch: "After reducing this dimension, don't remove it. Keep it, but make its size equal to 1."
```

```python
# 6.5 What is M.mean()?

TENSOR_1.float().mean()
```

```python
# 6.6 What is M.max(dim=1) – give both the values and the indices (argmax) along dim 1?

print(TENSOR_1.max()) # find the largest value
print(TENSOR_1.max(dim=1)) # returns maximum value in each row with its index.
```

```python
# What is M.argmax() on the flattened tensor (single index into the flattened array)?

print(TENSOR_1.reshape(-1).argmax()) # returns the index of the max element in complete tensor.
```

```python
# 6.8 What is M.var(dim=0, unbiased=False) (population vari-ance per column)?

print(TENSOR_1.float().mean())
print(TENSOR_1.float().var()) # compare with mean -> the far from mean the spread out values are
```
