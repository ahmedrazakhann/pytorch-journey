# Matrix Multiplication & Linear Algebra

```python
A = torch.tensor([[1.,2.],[3.,4.]])
B = torch.tensor([[5.,6.],[7.,8.]])
v = torch.tensor([1.,1.])

print(A)
print(B)
print(v)
```

```python
# 7.1 what is A @ B (matrix product)

A @ B
```

```python
# 7.2 What is A ⊙ B (elementwise product, NOT matmul)?
A * B
```

```python
# 7.3 What is A**T (transpose)?

A.transpose(0,1) # row into coloums OR cols into row.
```

```python
# 7.4 What is A @ v for v = [1, 1]?
print(A.shape)
print(v.shape)

# Matrix multiplication rule:
# (m×n)@(n×p)=(m×p)

A @ v
```

```python
# 7.5 What is det(A)?

torch.det(A) # formulae of determinant is cross multiply the matrix then subtract.
```

```python
# 7.6 What is A**−1 ?
print(A)
A.pow(-1)
```

```python
# 7.7 What is trace(A) (sum of diagonal)?

torch.trace(A)
# torch.trace(A.reshape(-1)) # will not work, it works only on square matrix as diagonals are there
```

```python
# 7.8 Given a batch of matrices X of shape (16, 4, 5) and Y of shape (16, 5, 3), what is the shape of torch.bmm(X, Y)

TENSOR_1 = torch.rand(16,4,5)
TENSOR_2 = torch.rand(16,5,3)

# print("TENSOR_1", TENSOR_1)
# print("TENSOR_2", TENSOR_2)

torch.bmm(TENSOR_1,TENSOR_2) # do matrix multiplcation for 3d tensors
```

```python
# Given X of shape (4, 5) and Y of shape (5, 3), what is the shape of X @ Y?

TENSOR_3 = torch.rand(4,5)
TENSOR_4 = torch.rand(5,3)

(TENSOR_3 @ TENSOR_4).shape # (4,3) -> *(m,n) x (n,p) = (m,p)
```

```python
# 7.10 Given X of shape (2, 4, 5) and a single matrix Y of shape
# (5, 3), what is the shape of X @ Y, and what rule allows
# this without bmm?



TENSOR_5 = torch.rand(2,4,5)
TENSOR_6 = torch.rand(5,3)


(TENSOR_5 @ TENSOR_6).shape # (2,4,3)

```
