# Elementwise & Arithmetic Operations

```python
x = torch.tensor([1.0,-2.0,3.0,-4.0])
y = torch.tensor([2.0,2.0,2.0,2.0])
```

```python
# 5.1 Given a1 = [1, −2, 3, −4], what is a.abs()?


a1 = torch.tensor([1, -2, 3, -4])
a1.abs()
```

```python
# 5.2 What is x2 (elementwise square)?

print(x ** 2)
```

```python
# 5.3 What is ReLU(x) = max(x, 0) elementwise?
import torch.nn.functional as F


print(F.relu(x))

# this function keep positive value and convert negative to 0
# --- simple analogy
#     - If positive = keep it
#     - If negative = replace with 0
```

```python
# 5.4 What is torch.sign(x)?

torch.sign(x)
```

```python
# What is x ⊙ y (elementwise product) for a2 = [1, −2, 3, −4], a3 = [2, 2, 2, 2]?

a2 = torch.tensor([1, -2, 3, -4])
a3 = torch.tensor([2, 2, 2, 2])

print(a2 * a3) 
```

```python
# 5.6 What is x/y elementwise?

print(x/y)
```

```python
# 5.7 What is torch.clamp(x, min=-1, max=2)?

torch.clamp(x, min=-1, max=3)
```

```python
# 5.8 What is the sigmoid σ(x) = 1/(1 + e−x ) applied element-wise to x = [1, −2, 3, −4]? Round to 4 decimals.

a4 = torch.tensor([1, -2, 9, -4])
torch.sigmoid(a4) # it map single input value into a smooth number between 0 - 1
```

```python
# 5.9 What is tanh(x) for x = [1, −2, 3, −4]?

torch.tanh(x) # it map single input value into smooth number between -1 and 1.
```

```python
# 5.10 What is torch.exp(x) for x = [1, −2, 3, −4]?

torch.exp(x) # exp(x)=e ** x
```

```python
# 5.11 Explain the difference between x.add(1) and x.add_(1) in PyTorch.
print(x)
print(x.add(1)) # it creates a new tensor, leaving unchanged the original tensor
print(x.add_(1)) # it updates the existing tensor
```

```python
# 5.13 What is x%3 (elementwise modulo) for integer tensor [7, 8, 9, 10]?
print(x)
print(x%3)
```

```python
# 5.14 What is torch.pow(x, 3) for x = [1, −2, 3, −4]?


torch.pow(x, 2) # its power on each element
```

```python
# 5.15 What is the result and dtype of torch.tensor([1,2,3]) / torch.tensor([2,2,2]) in PyTorch (true division vs integer division)?

a6 = torch.tensor([1,2,3]) / torch.tensor([2,2,2])
a6.dtype
```
