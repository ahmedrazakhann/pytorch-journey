```python
# Why does calling .view() sometimes raise a runtime error while .reshape() does not on the same tensor? 

# .view() only creates a new view of the same memory and requires the tensor to be contiguous.
# .reshape() tries to create a view first; if that is not possible because the tensor is non-contiguous, it creates a copy with new memory.
```

```python
# After y = B.permute(1,0,2), is y contiguous in memory? What must you call before you can safely .view() it?

y = B.permute(1,0,2)
y.is_contiguous() # results in FALSE
y = y.contiguous() # THIS HELPS IN STORING IT BACK TO CONTIGOUS MEMORY -> allocates new memorythe
y.is_contiguous() # results in TRUE

y.view(3,8) # now it would works
```
