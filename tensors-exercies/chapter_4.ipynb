```python
# 4.1 Can a tensor of shape (3, 4) be broadcast with one of shape (4, )? If so, what is the result shape?

# A: (3,4)
# B: (1,4)

# C: (3,4)

A = torch.rand(3,4)
B = torch.rand(4,)

C = A + B
C.shape
```

```python
# Can a tensor of shape (3, 1) be broadcast with one of shape (1, 4)? What is the result shape?

# Answer is yes it can broadcast, as it upper right dimension is one and could become 4 when stretched.

A1 = torch.rand(3,1)
A2 = torch.rand(1,4)

A3 = A1 * A2
A3
```

```python
# 4.3 Can shapes (3, 4) and (3, ) be broadcast together? Justify your answer using the broadcasting alignment rule.

# No it can't be broadcast as when we align the second tensor by moving it items to right so above 4 and bottom 3 don't match.  
# Broadcast rules says they both must match.
```

```python
# 4.4 What is the result shape of adding a tensor of shape (5, 1, 3) and one of shape (1, 4, 3)?

# it would be (5,4,3)
A4 = torch.rand(5,1,3)
A5 = torch.rand(1,4,3)

A6 = A4 + A5
A6.shape # here one would be replaced to the number according to the broadcasting rule.
```

```python
# 4.6 Given x of shape (N, D) and b of shape (D, ), what does x + b compute conceptually, and what is a common name for this pattern in neural networks?

# x + b adds the same bias vector b to every row/sample of x through broadcasting. 
# This is called bias broadcasting or bias addition, commonly used in neural network layers.
```

```python
# 4.9 Given an image batch of shape (N, 3, H, W ) and per-channel mean of shape (3, 1, 1), explain how normalization (images - mean) broadcasts.

# A tensor of shape (3,1,1) stores one mean value per channel. During images - mean, broadcasting expands it across the batch dimension N and spatial dimensions H, W, so each pixel in each channel has its corresponding channel mean subtracted. The output remains (N,3,H,W).
```
