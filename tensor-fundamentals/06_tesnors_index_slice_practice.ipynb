## Given the tensor above, write the slicing code to:

Print the first 3 rows.
Print the last 2 rows.
Print only the first column.
Print only the last column.
Print the top-left 2×2 block.
Print the bottom-right 2×2 block.
Print every second row.
Print every second column.
Print everything except the first row.
Print everything except the last column

```python
import torch

my_tensor = torch.arange(1,265, dtype=torch.int16)
print(my_tensor)
```

```python
print("Print the first 3 rows.")
print(my_tensor[0:3])
print(" \n")

print("Print the last 2 rows.")
print(my_tensor[6:8])
print(" \n")

print("Print only the first column.")
print(my_tensor[0,0])
print(" \n")

print("Print only the last column.")
print(my_tensor[7,7])
print(" \n")

print("Print the top-left 2×2 block.")
print(my_tensor[0:2,0:2])
print(" \n")

print("Print the bottom-right 2×2 block.")
print(my_tensor[6:8,6:8])
print(" \n")

print("Print every second row.")
print(my_tensor[::2])
print(" \n")

print("Print every second column.")
print(my_tensor[:,1])
print(" \n")

print("Print everything except the first row.")
print(my_tensor[1:])
print(" \n")

print("Print everything except the last row.")
print(my_tensor[:7])
```
