```python
import torch
print("Gpu enabled: ", torch.cuda.is_available())
```

```python
# addition
# rule: shape must be same

A = torch.arange(1,5).reshape(2,2)
B = torch.arange(5,9).reshape(2,2)

print(A)
print(B)
print(A + B)
print(torch.add(A,B))
```

```python
# subtraction
# rule: shape must be same

print(A)
print(B)
print(A - B)
print(torch.sub(A,B))
```

```python
# multiplication

tensor_a = torch.tensor([[2,2],[4,4]])
tensor_b = torch.tensor([[1,2],[3,4]])

# simple multiplication element-wide
print(tensor_a)
print(tensor_b)

print("Element wise mutiplication")
print(tensor_a * tensor_b)


print("Algebric multiplication")
print(tensor_a @ tensor_b) # OR
print(torch.matmul(tensor_a,tensor_b)) # OR
print(torch.mm(tensor_a,tensor_b)) # all are same


```

```python
# division
print(A)
print(B)
print(A / B)
print(torch.div(A,B))
```

```python
# power on matrix (on each element)
print(A ** 2)
```
