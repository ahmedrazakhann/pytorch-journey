# Reduction Operations

```python
TENSOR_1 = torch.randint(1,10,(3,2,3))
TENSOR_1
```

```python
# tensor.sum()

print(TENSOR_1)
print(TENSOR_1.sum()) # sum all values of a tensor
print(TENSOR_1.sum(dim=0)) # Sum across rows (vertical). One sum per column.
print(TENSOR_1.sum(dim=1, keepdim=True))   # Sum across columns (horizontal). One sum per row.

# keepdim=True tells PyTorch: "After reducing this dimension, don't remove it. Keep it, but make its size equal to 1."
```

```python
# 6.5 What is M.mean()?

TENSOR_1.float().mean()
```

```python
# 6.6 What is M.max(dim=1) – give both the values and the indices (argmax) along dim 1?

print(TENSOR_1.max()) # find the largest value
print(TENSOR_1.max(dim=1)) # returns maximum value in each row with its index.
```

```python
# What is M.argmax() on the flattened tensor (single index into the flattened array)?

print(TENSOR_1.reshape(-1).argmax()) # returns the index of the max element in complete tensor.
```

```python
# 6.8 What is M.var(dim=0, unbiased=False) (population vari-ance per column)?

print(TENSOR_1.float().mean())
print(TENSOR_1.float().var()) # compare with mean -> the far from mean the spread out values are
```

```python
# 6.9 What is M.std(dim=0, unbiased=False)?

TENSOR_1.float().std(dim=0, unbiased=False)
```

```python
# 6.10 What is M.prod(dim=1) (product along dim 1)?

print(TENSOR_1)
TENSOR_1.prod(dim=0)
```

```python
# 6.11 What is M.cumsum(dim=1)?

print(TENSOR_1.shape)
TENSOR_1.cumsum(dim=1)
```

```python
# 6.12 Explain the difference in output shape between M.sum(dim=1) and M.sum(dim=1, keepdim=True).

# keepdim=True keeps the reduced dimension in the output by making its size 1. Without keepdim, the reduced dimension is removed entirely.print(TENSOR_1)

print(TENSOR_1.shape)
print(TENSOR_1.sum(dim=1))
print(TENSOR_1.sum(dim=1).shape)
print(TENSOR_1.sum(dim=1, keepdim=True))
print(TENSOR_1.sum(dim=1, keepdim=True).shape)
```

```python
# 6.14 What is the result of torch.all(M > 0) and torch.any(M > 5)?

# all = Everyone must pass.
print(TENSOR_1 > 0) # compares with each element and return
print(torch.all(TENSOR_1 > 0)) # wants all elements to match the condition
print(torch.all(TENSOR_1 > 5))

# any = At least one must pass.
print(torch.any(TENSOR_1 > 5))
```

```python
# 6.13 What is torch.logsumexp(M, dim=1), and why is it nu-
# merically preferable to computing torch.log(torch.sum(torch.exp(M
# dim=1)) directly?



print(torch.logsumexp(TENSOR_1, dim=1))
# is basically:
print(torch.log(torch.sum(torch.exp(TENSOR_1), dim=1)))
```
