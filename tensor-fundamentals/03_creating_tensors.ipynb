```python
import torch
print("Gpu enabled: " , torch.cuda.is_available())
```

```python
a = torch.Tensor(2,4)
print("A: ", a)

b = torch.zeros(2, dtype=torch.int32)
print("B: ", b)

c = torch.ones(3, dtype=torch.int16)
print("C: ",c)

d = torch.arange(10,21)
print("D: ", d)

e = torch.linspace(start =-10, end = 10, steps = 3, dtype=torch.int32)
print("E: ", e)

f = torch.rand(3,4)
print("F: ", f)

g = torch.randn(2,3)
print("G: ",g)

e = torch.eye(2)
print("E: ", e)
```

```python
# tensors properties

print("A Shape: ", b.shape)
print("A Size: ", b.size())
print("A DType: ", b.dtype)
print("A Elements size: ", b.element_size())
print("_____________________________________________________")
print("A Dimension:" , b.ndim)
print("A Device: ", b.device)
print("A numelement: ", b.numel())
print("A layout: ", b.layout)
print("A requires_grad: ", b.requires_grad)
```
