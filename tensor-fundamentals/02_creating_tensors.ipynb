```python
import torch
print(f" Gpu enabled: {torch.cuda.is_available()}")
```

```python
a = torch.Tensor(2,3)
print("A: ", a)

b = torch.empty(2,3, dtype=torch.int32)
print("B: ", b)

c = torch.ones(2,3, dtype=torch.int64)
print("C: ", c)

d = torch.arange(4)
print("D: ", d)

e = torch.rand(3)
print("E: ", e)

f = torch.randn(3)
print("F: ", f)

g = torch.zeros(1,5,5, dtype=torch.int64)
print("G: ", g)

h = torch.zeros(1000,5,5, dtype=torch.int32)
print("H: ", h)

bytes_used = h.nelement() * h.element_size()
print(f"Space used: {bytes_used} bytes")
```

```python
tensor_1 = torch.zeros(1,2,2, dtype=torch.int16)
print("Tensor: ", tensor_1)
print("Device: ", tensor_1.device)
print("Shape: ", tensor_1.shape)
print("DType: ", tensor_1.dtype)
print("Dimensions: ", tensor_1.dim())
```
