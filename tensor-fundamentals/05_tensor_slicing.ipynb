```python
import torch
print("Gpu enabled: ", torch.cuda.is_available())
```

```python
# in this notebook we are going to cover tensors slicing, previously we learn tensors indexing that was about selecting one complete row/coloumn OR any specific element 
# inside it. 

# Tensors slicing is used for selecting range of elements.
```

```python
tensor_1 = torch.arange(1,65).reshape(8,8)
print(tensor_1)
```

```python
# ":" this colon is used for slicing but syntax is a bit different compare with indexing. 

# start : stop : step (this how its works lets understand in bit deep)

# start -> where to begin
# stop -> where to stop
# step -> how many positions to move each time

# a[rows, coloumn] 
```

```python
print(tensor_1)

print(" \n")
print("Start from second row (first index) and stop before 3rd row (till second row)")
print(tensor_1[1:3]) 

print(" \n")
print("Start from second row (skip first one)")
print(tensor_1[1:]) 

print(" \n")
print("Stop before 3rd row (second index)")
print(tensor_1[:2]) 
```

```python
print(tensor_1[:]) # complete row, cols
print(tensor_1[:,:2]) # this is index X slicing use means all rows and index 0,1 cols (stop before 2)

print(tensor_1[1:3])
```

```python
# Slicing always operates on the dimension you're slicing, means it varies.

# In simple words vector slicing works based on elements 1:3 from 1 before 3 right but in 2d tensor it works like 0 means first row not element so 1:3 means from 
# 1 row before 3rd row
```

```python
print(tensor_1[::])
print(tensor_1[1:3,0:3]) # two rows, 3 coloumns
print(tensor_1[1:3,2:]) # two rows, start at coloumn 2 till end
print(tensor_1[::3]) # ::3 -> the number is integer it tells that move this integer way on every index so on.
```

```python
print(tensor_1[2:]) # from 2 onwards
print(tensor_1[:2]) # all before 
print(tensor_1[::])
```

# pattern to understand

a[:]        # Select all rows (and all columns for a 2D tensor)
a[:2]       # Select rows before index 2 (rows 0 and 1)
a[2:]       # Select rows from index 2 onwards
a[:, :2]    # Select all rows, first 2 columns (columns 0 and 1)
a[:, 2:]    # Select all rows, columns from index 2 onwards
a[1:3, :]   # Select rows 1 and 2, all columns
a[:, 1:3]   # Select all rows, columns 1 and 2
