```python
a = torch.randint(0,10,(3,2))
print(a)
print(a[1,0]) # first index is for row and second is for coloumn
print(a[-1,0]) # here minus index indicates from bottom -1 = last row
print(a[:,0].reshape(3,1)) # print complete coloum
print(a[:,1]) # print second coloumn (first index) all rows
print(a[0,:]) # print first row (zero index) all coloumns OR simply a[0]
```

```python
b = torch.arange(1,10)
print(b)
print(b[1:4]) # this is known as range slicing that means b[start_index:end_index] give me value from start_index to end_index

c = torch.arange(10,19).reshape(3,3)
print(c)
```

```python
# at the end tensors indexing is used for selecting one specific element from any row/col a[0,2] = first row, second coloumn OR selecting any row complete a[3] or a[3,:] 
# means 4 row (3rd index) and all coloumns OR select only an coloumn using indexing.
```
